```python
import matplotlib.pyplot as plt
import matplotlib.patches as Circle

# Categories with compact labels
groups = {
    "Air Temp": ["Air T", "0.5 m T", "1.5 m T", "3 m T"],
    "Humidity": ["RH", "Rain", "Acc Rain"],
    "Radiation": ["Solar Rad"],
    "Wind": ["Wind Spd", "Wind Dir", "Wind Gust"],
    "Soil": [
        "4 in Bare T", "4 in Grass T", "2 in T", "2 in WC",
        "4 in T", "8 in T", "8 in WC", "20 in T", "20 in WC"
    ],
    "Stability": ["Inv Strength", "Max Inv"],
    "Other": ["Heat Index", "Wind Chill", "Pressure"]
}

# Circle positions and sizes
positions = {
    "Air Temp": (0.2, 0.7),
    "Humidity": (0.5, 0.8),
    "Radiation": (0.8, 0.7),
    "Wind": (0.15, 0.35),
    "Soil": (0.5, 0.35),
    "Stability": (0.85, 0.35),
    "Other": (0.5, 0.12)
}

sizes = {
    "Air Temp": 0.18,
    "Humidity": 0.17,
    "Radiation": 0.14,
    "Wind": 0.18,
    "Soil": 0.25,
    "Stability": 0.16,
    "Other": 0.14
}

colors = [
    "#82c8a0",
    "#ffd280",
    "#9fc5e8",
    "#f4a6b1",
    "#d5a6bd",
    "#c6b5e9",
    "#b6ccc9"
]

fig, ax = plt.subplots(figsize=(12, 8))
ax.set_aspect("equal")
ax.axis("off")

# Draw groups as circles
for idx, (group, vars) in enumerate(groups.items()):
    x, y = positions[group]
    r = sizes[group]

    circ = plt.Circle((x, y), r, color=colors[idx], alpha=0.45, ec="black", lw=1.3)
    ax.add_patch(circ)

    # Title text
    ax.text(x, y + r*0.4, group, ha="center", fontsize=13, weight="bold")

    # Variables arranged vertically inside each circle
    text = "\n".join(vars)
    ax.text(x, y, text, ha="center", va="center", fontsize=11)

plt.title("Variables Included in the Study", fontsize=18)
plt.tight_layout()
plt.show()
```
